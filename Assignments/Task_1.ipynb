### Q2:- Write a program that will convert celsius value to fahrenheit.

````python
# Write your code here
celsius=int(input("Enter temperature in celsius:"))
def celtofahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
print(celtofahrenheit(celsius))
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
import math
coord_1=[int(x) for x in input("enter coordinate-1:").split(',')]
coord_2=[int(x) for x in input("enter coordinate-2:").split(',')]
print(f"Coordinates:{coord_1},{coord_2}")
def euclidean_dis(c1,c2):
  return math.sqrt((c1[0] -c2[0])**2 + (c1[1]-c2[1])**2)
print(f"Distance between point-1:{coord_1} and point-2:{coord_2} is:{euclidean_dis(coord_1,coord_2)}")
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
def count(heads, legs, dogs=0):
    # Base case: if all heads are chickens
    if heads == legs // 2:
        chickens = heads
        return chickens, dogs

    # Recursive case: assume 1 dog, reduce heads & legs
    if heads < legs // 2:
        return count(heads - 1, legs - 4, dogs + 1)

    return None  # no solution

# ---- Main program ----
heads = int(input("Enter number of heads: "))
legs = int(input("Enter number of legs: "))

result = count(heads, legs)
if result:
    chickens, dogs = result
    print(f"Chickens: {chickens}, Dogs: {dogs}")
else:
    print("No valid solution!")

````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
n=int(input("Enter n:"))
sum=0
for x in range(1,n+1):
  sum+=pow(x,2)
print(f"Sum of square first {n} is:{sum}")
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here
f1=[int(x) for x in input("Enter fraction").split('/')]
f2=[int(x) for x in input("Enter fraction").split('/')]
sum= (f1[0]*f2[1]+f2[0]*f1[1])/(f1[1]*f2[1])
print(f"Sum of {f1} and {f2} is {sum}")
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
pi=3.14
H=int(input("Enter height of tank:"))
L=int(input("Enter length of tank:"))
B=int(input("Enter breadth of tank:"))
h=int(input("Enter height of glass:"))
r=int(input("Enter radius of glass:"))
no_glasses=(H*L*B)//(pi*r*r*h)
print("No. of glasses {}".format(no_glasses))
````
