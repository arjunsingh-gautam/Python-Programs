Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash

```

````python
# Write code here
num=int(input("Enter a number:"))
fact=1
if num==0 or num==1:
    fact=1
else:
    for x in range(1,num+1):
        fact=fact*x

print(f"Factorial of {num} is {fact}")
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# Write code here
n=int(input("Enter a number:"))
sum=0
for x in range(1,n+1):
    if x%5==0:
        print(f"Skipping number {x} divisible by 5")
        continue
    sum+=x
    if sum>300:
        print(f"Sum is greater than 300:{sum}")
        break
else:   
    print(F'Sum of first {n} numbers is:{sum}')
````

### `Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# Write code here
import math
current_position=[0,0]
start_position=[0,0]
while True:
    command=input("Enter the command:")
    if "up" in command.lower():
        move=int(command.split(' ')[-1])
        current_position[0]=current_position[0]+move
    elif "down" in command.lower():
        move=int(command.split(' ')[-1])
        current_position[0]=current_position[0]-move
    elif "left" in command.lower():
        move=int(command.split(' ')[-1])
        current_position[1]=current_position[1]-move
    elif "right" in command.lower():
        move=int(command.split(' ')[-1])
        current_position[1]=current_position[1]+move
    else:
        break

distance_travelled=int(math.sqrt((current_position[0]-start_position[0])**2+(current_position[1]-start_position[1])))
print(f"Distance travelled by robot between {start_position} and {current_position} is:{distance_travelled}")
````

### `Problem 12`:Write a program to print whether a given number is a prime number or not

````python
# Write code here
num=int(input("Enter a number"))
def isprime(num):
    if num==2:
        print(f"2 is prime number")
        return
    for i in range(2,num//2+1):
        if num%i==0:
            print(f"{num} not a prime number")
            break
    else:
        print(f"{num} is a prime number")
isprime(num)
````

### `Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
# Write code here
l=[]

n=int(input("Enter a number:"))
for i in range(n+1):
    y=0
    temp=i
    while temp>0:
        x=temp%10
        y+=pow(x,3)
        temp=temp//10
    if i==y:
        l.append(y)

print(f"Armstrong Number in range:0,{n} are:{l}")
````

### `Problem 15`:Given two rectangles, find if the given two rectangles overlap or not. A rectangle is denoted by providing the x and y coordinates of two points: the left top corner and the right bottom corner of the rectangle. Two rectangles sharing a side are considered overlapping. (L1 and R1 are the extreme points of the first rectangle and L2 and R2 are the extreme points of the second rectangle).

Note: It may be assumed that the rectangles are parallel to the coordinate axis.

<img src='https://www.geeksforgeeks.org/wp-content/uploads/rectanglesOverlap.png' width='300' height='200'>

````python
# Write code here
def do_rectangles_overlap(L1, R1, L2, R2):
    # Each point is a tuple (x, y) where x is horizontal and y is vertical.
    # Two rectangles do NOT overlap if one is to the left of the other or one is above the other.
    # L = left-top, R = right-bottom
    # If one rectangle is to the left of the other
    if R1[0] < L2[0] or R2[0] < L1[0]:
        return False
    # If one rectangle is above the other (note: y increases upward in this convention),
    # but many problems use the screen convention where y increases downward. We'll assume Cartesian coords where top has larger y than bottom.
    if L1[1] < R2[1] or L2[1] < R1[1]:
        return False
    return True

# Helper to parse input like: x1,y1 x2,y2 (space separated for two points)
def parse_point(s):
    x_str, y_str = s.split(',')
    return (int(x_str.strip()), int(y_str.strip()))

# Example interactive driver
print('Enter coordinates for first rectangle (L1 and R1). Use format x,y without spaces (e.g. 0,10)')
l1 = parse_point(input('L1 (left-top): '))
r1 = parse_point(input('R1 (right-bottom): '))
print('Enter coordinates for second rectangle (L2 and R2)')
l2 = parse_point(input('L2 (left-top): '))
r2 = parse_point(input('R2 (right-bottom): '))

if do_rectangles_overlap(l1, r1, l2, r2):
    print('Rectangles Overlap')
else:
    print('Rectangles Do Not Overlap')
````
