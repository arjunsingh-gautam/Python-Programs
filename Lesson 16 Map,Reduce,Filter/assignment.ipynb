## Q1.Double all even numbers


````python
nums=[1,2,3,4,5,6]
print(list(map(lambda x:x*2,list(filter(lambda x:True if x%2==0 else False,nums)))))
````

## Q2. Convert List of strings to integer:


````python
data=["10","20","30"]
int_data=list(map(lambda x:int(x),data))
print(int_data)
for x in int_data:
    print(f"Type of {x} is:{type(x)}")
````

## Q3. Sum of Squares

````python
nums=[1,2,3,4]
from functools import reduce
square=list(map(lambda x:x**2,nums))
square_sum=reduce(lambda x,y:x+y,square)
print(square_sum)
````

## Q4. Create a Dictionary from two list

## Q5. Printed ranked scores

````python
scores=[88,92,79]
for i,score in enumerate(scores,start=1):
    print(f"Rank {i}: {score}")
````
